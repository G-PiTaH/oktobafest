TUSKER OKTOBAFEST ANALYSIS

```python
#IMPORTING TOOLS
import pandas as pd   #Tool used for dataframe manipulation
import snscrape.modules.twitter as sntwitter # Tool to scrape tweets
```

Data Gathering from tweets

```python
#Defining search query
search_query = ' "since:2022-09-02" #TuskerOktobafest OR #TuskerOktobafestStepsChallenge OR Octoberfest'
# english tweets as NLP is tobe used later for tokenization
tweets = [] #List to append scraped tweets initially empty
limit = 41304 #Number of tweets inspired by total tusker oktobafest attendance
count = 0
```

```python
for tweet in sntwitter.TwitterSearchScraper(search_query).get_items():
    count += 1 # to see progress as the tweets scrape to prevent insanity 
    # print(vars(tweet))  # printing the variables to see tweet's attributes
    # break
    if len(tweets) == limit:
        break
    else:
        # append necessary attributes to empty list
        tweets.append([tweet.date, tweet.content, tweet.user.username, 
        tweet.user.location, tweet.retweetCount, tweet.likeCount, tweet.sourceLabel, tweet.coordinates])
        print('Number ',count, ':success')

# turn 'tweets' list into a dataframe
df = pd.DataFrame(tweets, columns=['date','text','username','location','retweetcount','likecount','source','coordinates'])

# convert dataframe into csv file
df.to_csv('OKTOBAFEST.csv', index=False)
```

Data Cleaning

```python
df_tof['date'] = pd.to_datetime(df_tof['date']) #selecting datatype for the date
```

```python
import re #For expressions
```

```python
def cleanText(text):
    text = text.lower()
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)

    text = re.sub('\@\w+|\#\w+|\d+','',text) # Remove @mentions
    text = re.sub('http(s?)(:?)\/\/\S+','', text) # Removing the hyper link
    text = re.sub('#','',text)  # Removing the # symbols
    text = re.sub(':','',text)  # Removing the : symbols
    text = re.sub('\"','',text) # Removing the " symbols
    text = re.sub('\','',text) # Removing the  symbols
    text = re.sub('\','',text) # Removing the  symbols
    text = re.sub('\','',text) # Removing the  symbols
    text = re.sub('\/','',text) # Removing the / symbols
    text = re.sub('\|','',text) # Removing the | symbols
    text = re.sub('\?','',text) # Removing the ? symbols
    text = re.sub('\.','',text) # Removing the . symbols
    text = re.sub('\,','',text) # Removing the , symbols
    text = re.sub('\!','',text) # Removing the ! symbols
    text = re.sub('@','',text) # Removing the @ symbols
    text = re.sub('\'','',text) # Removing the ' symbols
    text = re.sub('\\n','',text) # Removing the \n symbols
    text = re.sub('\\r','',text) # Removing the \r symbols
    text = re.sub('\*','',text) # Removing the * symbols
    text = emoji_pattern.sub('', text) # Removing emoji patterns
    return text

df_tof['cleaned_text'] = df_tof['text'].apply(cleanText)
```
